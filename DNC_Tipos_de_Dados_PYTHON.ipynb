###Tipos de Dados
Tipos de dados

Python possui alguns tipos de dados primitivos que são muito utilizados:


String: cadeia de caracteres entre aspas simples ou duplas. Pode conter texto, números, caracteres especiais etc. Exemplos: 'Olá Mundo', "10", "#$ab".

Int: números inteiros. Exemplos: 5, 10, -20.

Float: números de ponto flutuante, com casa decimal. Em Python o separador de decimal é o ponto, não a vírgula. Exemplos: 10.5, 3.1415, -15.8.

Boolean: valores booleanos Verdadeiro ou Falso. São representados por True e False (sempre com inicial maiúscula e sem aspas).


Funções básicas

Algumas funções básicas do Python que foram apresentadas:

Print

A função print() é usada para imprimir, exibir algum valor na tela. Podemos printar strings, números, variáveis etc.

print("Olá mundo")
print(5)
print(minha_variavel)


Dentro do print podemos colocar múltiplos valores separados por vírgula, que serão exibidos em sequência.

print("Meu nome é", nome, "e eu tenho", idade, "anos")


Input

A função input() permite ler um valor digitado pelo usuário e armazenar em uma variável. Sempre retorna strings.

lanche = input("Qual lanche você deseja?")
# Exibe "Qual lanche você deseja?" e espera o usuário digitar


Type

Retorna o tipo de dado de um valor ou variável. Útil para fazer validações.

type("Maria") # Retorna string
type(10) # Retorna int
type(8.5) # Retorna float


Variáveis

Variáveis são usadas para armazenar valores na memória para reutilização durante o programa.

Em Python declaramos variáveis simplesmente atribuindo valores a elas com o sinal de =.

# Atribuições
nome = "João"
idade = 30
pi = 3.1415
solteiro = True


Para declarar, basta especificar o nome da variável, sinal de =, e o valor.

O valor pode ser passado diretamente ou ser resultado de uma expressão ou cálculo.

preco_maca = 2.30
qtde = 10
total = preco_maca * qtde


Uma vez declaradas, podemos utilizar as variáveis em nosso código no lugar de valores literais.

print("O cliente", nome, "comprou", qtde, "maçãs")
print("O total é R$", total)


Entrada de dados pelo usuário

Podemos pedir para o usuário informar valores durante a execução utilizando a função input(), que sempre retorna strings.

qtde = input("Quantas maçãs você deseja?")
nome = input("Qual o seu nome?")


Operadores aritméticos

Podemos utilizar operadores aritméticos em Python para fazer contas:


Soma com +
Subtração com -
Multiplicação com *
Divisão com /
Divisão inteira com //
Resto da divisão com %
Exponenciação com **

soma = 10 + 15 #25
sub = 20 - 5 #15
mult = 10 * 2 #20
div = 20 / 4 #5
resto = 7 % 2 #1
pot = 2 ** 4 #16


Concatenação de strings

A concatenação une duas strings, colocando uma depois da outra. Em Python podemos fazer de duas formas:


Com o sinal de +:

nome = "João"
print("Olá " + nome)



Separando por vírgula no print:

```python
nome = "João"
print("Olá", nome) #Adiciona espaço entre as strings

A segunda forma adiciona um espaço entre as strings e é mais comum.

## Tomada de decisões e booleans

Variáveis booleanas (`True`, `False`) são muito utilizadas para tomar decisões dentro de programas.

Podemos atribuir resultados booleanos de comparações a variáveis:


python
idade = 15
maior_18 = idade >= 18 # False

cartao = True
pagou_cartao = cartao # True

E utilizar essas variáveis em estruturas condicionais e de repetição:


python
if maior_18:
print("Pode entrar na festa")
else:
print("Entrada proibida")

if pagou_cartao:
print("Pagamento aprovado")
else:
print("Pagamento recusado")

## Exemplos práticos

Vamos ver alguns exemplos práticos, passo a passo:

### Exemplo 1 - Olá mundo

Vamos imprimir a frase "Olá mundo":


python
print("Olá mundo")

Funciona com aspas simples ou duplas.

### Exemplo 2 - Operações matemáticas

Vamos fazer algumas contas simples:


python
print(5 + 7) # 12
print(10 - 3) # 7
print(4 * 5) # 20
print(10 / 2) # 5

Podemos colocar números diretamente ou variáveis previamente definidas.

### Exemplo 3 - Trabalhando com variáveis

Vamos declarar algumas variáveis:


python
nome = "Maria"
idade = 19
grana = 51.30

print(nome, "tem", idade, "anos e R$", grana, "na carteira")

Saída:


Maria tem 19 anos e R$ 51.3 na carteira

Perceba que usamos as variáveis dentro do print.

### Exemplo 4 - Input de dados

Vamos pedir para o usuário digitar seu nome e idade e exibir:


python
nome = input("Digite seu nome: ")
idade = input("Digite sua idade: ")

print(nome, "tem", idade, "anos")

Saída:


Digite seu nome: José
Digite sua idade: 28
José tem 28 anos

Note que precisamos armazenar em variáveis para reutilizar depois.

### Exemplo 5 - Concatenação

Vamos concatenar (juntar) alguns valores:


python
nome = "Bruna"
sobrenome = "Silva"

print(nome + " " + sobrenome) # Com +
print(nome, sobrenome) # Com vírgula

Saídas:


Bruna Silva
Bruna Silva

Perceba o espaço adicional na segunda frase.

### Exemplo 6 - Tipos de dados

Vamos verificar tipos de alguns dados:


python
type("Teste") # String
type(10) # Int
type(10.5) # Float
type(True) # Bool
```
