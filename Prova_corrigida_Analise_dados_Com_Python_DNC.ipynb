Questo 8 - O que o mtodo dropna() faz quando aplicado a um DataFrame?

O mtodo `dropna()` do **pandas**  usado para **remover linhas ou colunas que contenham valores ausentes (NaN)** em um DataFrame. Ele permite limpar o conjunto de dados eliminando entradas incompletas.

### Funcionamento:
- **Por padro**, `dropna()` remove as **linhas** que contm pelo menos um valor `NaN`.
- Ele tambm pode ser configurado para remover **colunas** com valores ausentes.

### Sintaxe:
```python
df.dropna(axis=0, how='any', inplace=False)
```

### Argumentos principais:
- **`axis=0`**: Remove **linhas** (comportamento padro). Se `axis=1`, remove **colunas**.
- **`how='any'`**: Remove linhas ou colunas que tenham **qualquer** valor `NaN`. Se `how='all'`, remove apenas se **todos** os valores na linha ou coluna forem `NaN`.
- **`inplace=False`**: Se for `True`, faz a modificao no DataFrame original. Se `False`, retorna um novo DataFrame sem os valores ausentes.

### Exemplo:
```python
import pandas as pd

data = {'A': [1, 2, None], 'B': [4, None, 6]}
df = pd.DataFrame(data)

# Remover as linhas com valores NaN
df_cleaned = df.dropna()

print(df_cleaned)
```

Neste exemplo, as linhas contendo `NaN` so removidas, resultando em um DataFrame limpo.

Questo 9 - O mtodo join() no Pandas  equivalente ao mtodo merge() quando se trata de combinar DataFrames. Esta afirmao ?

A afirmao de que o mtodo `join()` no **Pandas**  equivalente ao mtodo `merge()` quando se trata de combinar DataFrames  **parcialmente verdadeira**, pois eles tm propsitos semelhantes, mas funcionam de maneira um pouco diferente.

### Diferenas entre `join()` e `merge()`:

1. **`join()`**:
   -  usado principalmente para combinar DataFrames com base no **ndice** (por padro).
   - Ele pode combinar DataFrames de maneira semelhante a um **left join** por padro, mas pode ser ajustado para outros tipos de joins.
   - Sintaxe mais simples quando se deseja combinar DataFrames que j possuem ndices alinhados.
   
   Exemplo:
   ```python
   df1.join(df2, on='key', how='left')
   ```

2. **`merge()`**:
   -  mais flexvel e usado para combinar DataFrames com base em **colunas** (no apenas ndices).
   - Pode ser configurado para executar diferentes tipos de joins (`inner`, `outer`, `left`, `right`).
   - Permite combinar em uma ou mais colunas, sem exigir que as colunas sejam ndices.
   
   Exemplo:
   ```python
   pd.merge(df1, df2, on='key', how='left')
   ```

### Concluso:
A afirmao  **parcialmente verdadeira**. Ambos os mtodos combinam DataFrames, mas `join()`  mais limitado (focado em ndices), enquanto `merge()` oferece mais controle e flexibilidade, permitindo combinaes com base em colunas e mltiplos tipos de join.
