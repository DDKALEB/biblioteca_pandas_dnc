Questão 8 - O que o método dropna() faz quando aplicado a um DataFrame?

O método `dropna()` do **pandas** é usado para **remover linhas ou colunas que contenham valores ausentes (NaN)** em um DataFrame. Ele permite limpar o conjunto de dados eliminando entradas incompletas.

### Funcionamento:
- **Por padrão**, `dropna()` remove as **linhas** que contêm pelo menos um valor `NaN`.
- Ele também pode ser configurado para remover **colunas** com valores ausentes.

### Sintaxe:
```python
df.dropna(axis=0, how='any', inplace=False)
```

### Argumentos principais:
- **`axis=0`**: Remove **linhas** (comportamento padrão). Se `axis=1`, remove **colunas**.
- **`how='any'`**: Remove linhas ou colunas que tenham **qualquer** valor `NaN`. Se `how='all'`, remove apenas se **todos** os valores na linha ou coluna forem `NaN`.
- **`inplace=False`**: Se for `True`, faz a modificação no DataFrame original. Se `False`, retorna um novo DataFrame sem os valores ausentes.

### Exemplo:
```python
import pandas as pd

data = {'A': [1, 2, None], 'B': [4, None, 6]}
df = pd.DataFrame(data)

# Remover as linhas com valores NaN
df_cleaned = df.dropna()

print(df_cleaned)
```

Neste exemplo, as linhas contendo `NaN` são removidas, resultando em um DataFrame limpo.

Questão 9 - O método join() no Pandas é equivalente ao método merge() quando se trata de combinar DataFrames. Esta afirmação é?

A afirmação de que o método `join()` no **Pandas** é equivalente ao método `merge()` quando se trata de combinar DataFrames é **parcialmente verdadeira**, pois eles têm propósitos semelhantes, mas funcionam de maneira um pouco diferente.

### Diferenças entre `join()` e `merge()`:

1. **`join()`**:
   - É usado principalmente para combinar DataFrames com base no **índice** (por padrão).
   - Ele pode combinar DataFrames de maneira semelhante a um **left join** por padrão, mas pode ser ajustado para outros tipos de joins.
   - Sintaxe mais simples quando se deseja combinar DataFrames que já possuem índices alinhados.
   
   Exemplo:
   ```python
   df1.join(df2, on='key', how='left')
   ```

2. **`merge()`**:
   - É mais flexível e usado para combinar DataFrames com base em **colunas** (não apenas índices).
   - Pode ser configurado para executar diferentes tipos de joins (`inner`, `outer`, `left`, `right`).
   - Permite combinar em uma ou mais colunas, sem exigir que as colunas sejam índices.
   
   Exemplo:
   ```python
   pd.merge(df1, df2, on='key', how='left')
   ```

### Conclusão:
A afirmação é **parcialmente verdadeira**. Ambos os métodos combinam DataFrames, mas `join()` é mais limitado (focado em índices), enquanto `merge()` oferece mais controle e flexibilidade, permitindo combinações com base em colunas e múltiplos tipos de join.
