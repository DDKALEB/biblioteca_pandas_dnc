```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vendas = pd.read_excel('/content/varejo.xlsx')
```

```python
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique().sort_values(ascending=False)
```

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP", "Aplicativo")
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby("bandeira").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.groupby("estado").idcompra.nunique()
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ", "_")
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas[vendas["estado"].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna() # aqui tiramos  todos os nulos, mas escluimos dados que nao podemos
vendas_sem_nulos.info()
```

```python
vendas["estado"] .fillna("MS", inplace=True) #aqqui voltamos a pegar a base de dados original e substituimos os  nulos por MS
vendas.info()
```

Perceba agora que so teemos dados nulos para Precos !!

```python
media_preco = vendas["Preço"].mean() # aqui calculamos o valor medio de preço e substituimos
vendas["Preço"].fillna(media_preco, inplace=True)
media_preco
```

```python
vendas.info()
```

```python
mediana_preco = vendas["Preço"].median() # aqui calculamos o valor DA mediana de preço e substituimos - MEDIANA e o ponto medio que divide os dados para cia=ma e abaixo
vendas["Preço"].fillna(mediana_preco, inplace=True)
mediana_preco
```

```python
vendas.head(  )
```

```python
vendas[["Preço", "Preço_com_frete"]].describe()
```

```python
preco_errado = vendas.query("Preço > Preço_com_frete")
preco_errado
```

```python
vendas_correto = vendas.query("Preço < Preço_com_frete")
vendas_correto.describe()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP' ").describe()
```

```python
vendas_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index()
```

```python
round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
```

```python
vendas_correto.groupby("Data").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
vendas_correto["mes"] = vendas_correto["Data"].dt.month
vendas_correto.head()
```

```python
vendas_correto.groupby("mes").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
vendas_cliente = vendas_correto.merge(cliente, how="left", on="cliente_Log")
vendas_cliente.head()
```

```python
agg_idcanal_renda = vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").reset_index()
agg_idcanal_renda
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").reset_index(),2)
agg_idcanal_renda
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby(["bandeira", "idade"])["renda"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"], color="green")
plt.title("Média de renda por canal de venda", loc="center", fontsize=18, fontweight="bold", color="black");
plt.ylabel("Média de renda")
plt.xlabel("Canal de venda")
plt.show()
```
