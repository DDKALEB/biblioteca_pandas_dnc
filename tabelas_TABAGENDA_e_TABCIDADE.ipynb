```python
import sqlite3

# Conexão com o banco de dados
conexao = sqlite3.connect('DBAGENDA.db')
cursor = conexao.cursor()

# Criação da tabela TABAGENDA, caso não exista
cursor.execute('''CREATE TABLE IF NOT EXISTS TABAGENDA (
    CODAGENDA INTEGER PRIMARY KEY,
    NOME TEXT NOT NULL,
    TELEFONE TEXT,
    ENDERECO TEXT,
    CODCIDADEAGENDA INTEGER
)''')

# Criação da tabela TABCIDADE, caso não exista
cursor.execute('''CREATE TABLE IF NOT EXISTS TABCIDADE (
    CODCIDADE INTEGER PRIMARY KEY,
    NOMECIDADE TEXT NOT NULL
)''')

# Função de inclusão de registros
def incluir_registro():
    VARCODAGENDA = int(input("Digite o código do registro: "))
    VARNOME = input("Digite o nome do contato: ")
    VARTELEFONE = input("Digite o telefone: ")
    VARENDERECO = input("Digite o endereço: ")
    VARCODCID = int(input("Digite o código da cidade: "))

    # Verificar se o código já existe na TABAGENDA
    cursor.execute("SELECT * FROM TABAGENDA WHERE CODAGENDA = ?", (VARCODAGENDA,))
    if cursor.fetchone():
        print("Código já existente.")
        return

    # Verificar se o código da cidade existe
    cursor.execute("SELECT * FROM TABCIDADE WHERE CODCIDADE = ?", (VARCODCID,))
    if not cursor.fetchone():
        print("Código da cidade inválido. Cidade não existente.")
        return

    # Inserir o registro na TABAGENDA
    cursor.execute("INSERT INTO TABAGENDA (CODAGENDA, NOME, TELEFONE, ENDERECO, CODCIDADEAGENDA) VALUES (?, ?, ?, ?, ?)",
                   (VARCODAGENDA, VARNOME, VARTELEFONE, VARENDERECO, VARCODCID))
    conexao.commit()
    print("Registro incluído com sucesso.")

# Função de consulta de registros
def consultar_registro():
    VARNOME = input("Digite o nome do contato para buscar: ")

    cursor.execute("SELECT * FROM TABAGENDA WHERE NOME = ?", (VARNOME,))
    registros = cursor.fetchall()

    if registros:
        for registro in registros:
            CODAGENDA, NOME, TELEFONE, ENDERECO, CODCIDADEAGENDA = registro
            # Buscar nome da cidade
            cursor.execute("SELECT NOMECIDADE FROM TABCIDADE WHERE CODCIDADE = ?", (CODCIDADEAGENDA,))
            NOMECIDADE = cursor.fetchone()[0]
            print(f"Código: {CODAGENDA}, Nome: {NOME}, Telefone: {TELEFONE}, Endereço: {ENDERECO}, Cidade: {NOMECIDADE}")
    else:
        print("Contato não existente.")

# Função de alteração de registros
def alterar_registro():
    VARCODAGENDA = int(input("Digite o código do registro a ser alterado: "))

    cursor.execute("SELECT * FROM TABAGENDA WHERE CODAGENDA = ?", (VARCODAGENDA,))
    registro = cursor.fetchone()

    if registro:
        VARCODAGENDA, VARNOME, VARTELEFONE, VARENDERECO, VARCODCID = registro
        # Mostrar dados atuais
        print(f"Nome atual: {VARNOME}")
        print(f"Telefone atual: {VARTELEFONE}")
        print(f"Endereço atual: {VARENDERECO}")
        print(f"Código da cidade atual: {VARCODCID}")

        # Alterar dados
        VARNOME = input("Digite o novo nome (ou Enter para manter o atual): ") or VARNOME
        VARTELEFONE = input("Digite o novo telefone (ou Enter para manter o atual): ") or VARTELEFONE
        VARENDERECO = input("Digite o novo endereço (ou Enter para manter o atual): ") or VARENDERECO
        VARCODCID = int(input("Digite o novo código da cidade (ou Enter para manter o atual): ") or VARCODCID)

        # Atualizar registro
        cursor.execute("UPDATE TABAGENDA SET NOME = ?, TELEFONE = ?, ENDERECO = ?, CODCIDADEAGENDA = ? WHERE CODAGENDA = ?",
                       (VARNOME, VARTELEFONE, VARENDERECO, VARCODCID, VARCODAGENDA))
        conexao.commit()
        print("Registro alterado com sucesso.")
    else:
        print("Código não existente.")

# Função de exclusão de registros
def excluir_registro():
    VARCODAGENDA = int(input("Digite o código do registro a ser excluído: "))

    cursor.execute("SELECT * FROM TABAGENDA WHERE CODAGENDA = ?", (VARCODAGENDA,))
    registro = cursor.fetchone()

    if registro:
        CODAGENDA, NOME, TELEFONE, ENDERECO, CODCIDADEAGENDA = registro
        # Mostrar dados atuais
        print(f"Código: {CODAGENDA}, Nome: {NOME}, Telefone: {TELEFONE}, Endereço: {ENDERECO}")
        confirmacao = input("Deseja excluir este registro? (s/n): ").lower()
        if confirmacao == 's':
            cursor.execute("DELETE FROM TABAGENDA WHERE CODAGENDA = ?", (CODAGENDA,))
            conexao.commit()
            print("Registro excluído com sucesso.")
        else:
            print("Operação cancelada.")
    else:
        print("Código não existente.")

# Função para acessar a tabela de cidades
def acessar_tabela_cidades():
    print("Acessando a tabela de cidades...")
    cursor.execute("SELECT * FROM TABCIDADE")
    cidades = cursor.fetchall()
    for cidade in cidades:
        CODCIDADE, NOMECIDADE = cidade
        print(f"Código: {CODCIDADE}, Nome da Cidade: {NOMECIDADE}")

# Função principal para o menu
def menu():
    while True:
        print("\nMENU PRINCIPAL DA AGENDA")
        print("1. Incluir Registro")
        print("2. Consultar Registro")
        print("3. Alterar Registro")
        print("4. Excluir Registro")
        print("5. Tabela de Cidades")
        print("6. Sair")
        opcao = input("Selecione uma opção: ")

        if opcao == '1':
            incluir_registro()
        elif opcao == '2':
            consultar_registro()
        elif opcao == '3':
            alterar_registro()
        elif opcao == '4':
            excluir_registro()
        elif opcao == '5':
            acessar_tabela_cidades()
        elif opcao == '6':
            print("Saindo do sistema...")
            break
        else:
            print("Opção inválida. Tente novamente.")

# Executar o menu principal
menu()

# Fechar a conexão com o banco de dados ao sair
conexao.close()
```

```python
import sqlite3
from tkinter import *
from tkinter import messagebox
from tkinter import ttk

# Conexão com o banco de dados
conn = sqlite3.connect('DBAGENDA.db')
cursor = conn.cursor()

# Função para listar todos os registros da agenda
def listar_registros():
    janela_listagem = Toplevel()
    janela_listagem.title("Listar Registros")

    # Adiciona um Frame com Scrollbar
    frame = Frame(janela_listagem)
    frame.pack(fill=BOTH, expand=True)

    canvas = Canvas(frame)
    canvas.pack(side=LEFT, fill=BOTH, expand=True)

    scrollbar = ttk.Scrollbar(frame, orient=VERTICAL, command=canvas.yview)
    scrollbar.pack(side=RIGHT, fill=Y)

    scroll_frame = Frame(canvas)
    canvas.create_window((0, 0), window=scroll_frame, anchor="nw")

    # Busca todos os registros
    cursor.execute('''SELECT TABAGENDA.CODAGENDA, TABAGENDA.NOME, TABAGENDA.TELEFONE,
                      TABAGENDA.ENDERECO, TABCIDADE.NOMECIDADE
                      FROM TABAGENDA
                      JOIN TABCIDADE ON TABAGENDA.CODCIDADE = TABCIDADE.CODCIDADE''')

    registros = cursor.fetchall()

    # Exibe os registros na tela
    for i, registro in enumerate(registros):
        Label(scroll_frame, text=f"Código: {registro[0]}, Nome: {registro[1]}, Telefone: {registro[2]}, "
                                 f"Endereço: {registro[3]}, Cidade: {registro[4]}").grid(row=i, column=0, sticky="w")

    # Atualiza o scroll
    scroll_frame.update_idletasks()
    canvas.config(scrollregion=canvas.bbox("all"), yscrollcommand=scrollbar.set)

# Função para listar todas as cidades
def listar_cidades():
    janela_listagem_cidades = Toplevel()
    janela_listagem_cidades.title("Listar Cidades")

    # Adiciona um Frame com Scrollbar
    frame = Frame(janela_listagem_cidades)
    frame.pack(fill=BOTH, expand=True)

    canvas = Canvas(frame)
    canvas.pack(side=LEFT, fill=BOTH, expand=True)

    scrollbar = ttk.Scrollbar(frame, orient=VERTICAL, command=canvas.yview)
    scrollbar.pack(side=RIGHT, fill=Y)

    scroll_frame = Frame(canvas)
    canvas.create_window((0, 0), window=scroll_frame, anchor="nw")

    # Busca todas as cidades
    cursor.execute("SELECT * FROM TABCIDADE")
    cidades = cursor.fetchall()

    # Exibe as cidades na tela
    for i, cidade in enumerate(cidades):
        Label(scroll_frame, text=f"Código: {cidade[0]}, Cidade: {cidade[1]}").grid(row=i, column=0, sticky="w")

    # Atualiza o scroll
    scroll_frame.update_idletasks()
    canvas.config(scrollregion=canvas.bbox("all"), yscrollcommand=scrollbar.set)

# Interface principal
janela = Tk()
janela.title("Menu Principal da Agenda")

# Botões principais
Button(janela, text="Incluir", command=incluir_registro).grid(row=0, column=0)
Button(janela, text="Listar Registros", command=listar_registros).grid(row=1, column=0)
Button(janela, text="Listar Cidades", command=listar_cidades).grid(row=2, column=0)
Button(janela, text="Sair", command=janela.quit).grid(row=3, column=0)

janela.mainloop()

# Fechar conexão com o banco de dados ao sair
conn.close()
```

```python
import sqlite3
from tkinter import *

# Conexão com o banco de dados
conn = sqlite3.connect('agenda.db')
cursor = conn.cursor()

# Criação das tabelas TABAGENDA e TABCIDADE se não existirem
cursor.execute('''CREATE TABLE IF NOT EXISTS TABCIDADE (
                    CODCIDADE INTEGER PRIMARY KEY AUTOINCREMENT,
                    NOMECIDADE TEXT NOT NULL)''')

cursor.execute('''CREATE TABLE IF NOT EXISTS TABAGENDA (
                    CODAGENDA INTEGER PRIMARY KEY AUTOINCREMENT,
                    NOME TEXT NOT NULL,
                    TELEFONE TEXT NOT NULL,
                    ENDERECO TEXT NOT NULL,
                    CODCIDADE INTEGER,
                    FOREIGN KEY (CODCIDADE) REFERENCES TABCIDADE(CODCIDADE))''')

# Função para incluir um registro
def incluir_registro():
    janela_incluir = Toplevel()
    janela_incluir.title("Incluir Registro")

    Label(janela_incluir, text="Nome").grid(row=0, column=0)
    nome_entry = Entry(janela_incluir)
    nome_entry.grid(row=0, column=1)

    Label(janela_incluir, text="Telefone").grid(row=1, column=0)
    telefone_entry = Entry(janela_incluir)
    telefone_entry.grid(row=1, column=1)

    Label(janela_incluir, text="Endereço").grid(row=2, column=0)
    endereco_entry = Entry(janela_incluir)
    endereco_entry.grid(row=2, column=1)

    Label(janela_incluir, text="Código da Cidade").grid(row=3, column=0)
    codcidade_entry = Entry(janela_incluir)
    codcidade_entry.grid(row=3, column=1)

    def salvar():
        nome = nome_entry.get()
        telefone = telefone_entry.get()
        endereco = endereco_entry.get()
        codcidade = codcidade_entry.get()

        cursor.execute('''INSERT INTO TABAGENDA (NOME, TELEFONE, ENDERECO, CODCIDADE)
                          VALUES (?, ?, ?, ?)''', (nome, telefone, endereco, codcidade))
        conn.commit()
        janela_incluir.destroy()

    Button(janela_incluir, text="Salvar", command=salvar).grid(row=4, column=1)

# Função para listar todos os registros da agenda no console
def listar_registros_console():
    try:
        # Busca todos os registros
        cursor.execute('''SELECT TABAGENDA.CODAGENDA, TABAGENDA.NOME, TABAGENDA.TELEFONE,
                          TABAGENDA.ENDERECO, TABCIDADE.NOMECIDADE
                          FROM TABAGENDA
                          JOIN TABCIDADE ON TABAGENDA.CODCIDADE = TABCIDADE.CODCIDADE''')

        registros = cursor.fetchall()

        # Exibe os registros no console
        if registros:
            for registro in registros:
                print(f"Código: {registro[0]}, Nome: {registro[1]}, Telefone: {registro[2]}, "
                      f"Endereço: {registro[3]}, Cidade: {registro[4]}")
        else:
            print("Nenhum registro encontrado.")
    except sqlite3.OperationalError as e:
        print(f"Erro ao listar registros: {e}")

# Função para listar todas as cidades no console
def listar_cidades_console():
    try:
        cursor.execute("SELECT * FROM TABCIDADE")
        cidades = cursor.fetchall()

        # Exibe as cidades no console
        if cidades:
            for cidade in cidades:
                print(f"Código: {cidade[0]}, Cidade: {cidade[1]}")
        else:
            print("Nenhuma cidade encontrada.")
    except sqlite3.OperationalError as e:
        print(f"Erro ao listar cidades: {e}")

# Interface principal (caso haja suporte gráfico)
try:
    janela = Tk()
    janela.title("Menu Principal da Agenda")

    # Botões principais
    Button(janela, text="Incluir", command=incluir_registro).grid(row=0, column=0)
    Button(janela, text="Listar Registros", command=listar_registros_console).grid(row=1, column=0)
    Button(janela, text="Listar Cidades", command=listar_cidades_console).grid(row=2, column=0)
    Button(janela, text="Sair", command=janela.quit).grid(row=3, column=0)

    janela.mainloop()

except TclError:
    print("Ambiente gráfico indisponível. Usando modo de listagem no console.")
    listar_registros_console()
    listar_cidades_console()

# Fechar conexão com o banco de dados ao sair
conn.close()
```
